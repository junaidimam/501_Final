```python
mask_cols = ml_df['masking']
```

```python
len(mask_cols)
```

```python
pt_mask=[]
masking_level =[]
for val in mask_cols:
  if("Participant" in val):
    pt_mask.append("Yes")
  else:
    pt_mask.append("No")
  
  if("Quadruple" in val):
    masking_level.append(4)
  elif("Triple" in val):
    masking_level.append(3)
  elif("Double" in val):
    masking_level.append(2)
  elif("single" in val):
    masking_level.append(1)
  elif("None" in val):
    masking_level.append(0)
  else:
    masking_level.append(-1)
```

```python
len(masking_level)
```

```python
ml_df['pt_mask'] = pt_mask
ml_df['masking_level'] = masking_level
```

```python
ml_df = ml_df.drop(['masking'] , axis=1) 
```

```python
ml_df = ml_df.drop(['Unnamed: 0'] , axis=1)
```
