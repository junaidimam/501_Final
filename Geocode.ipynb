```python
zcdb = ZipCodeDatabase()
```

```python
lat_list =[]
long_list = []
error = []
zip_new = []
for index , row in df.iterrows():
  
  try:
    zip = (row['zip'][0:5])
    
    temp = zcdb[zip]
    lat = temp.latitude
    longi = temp.longitude
  
  except:
    error.append(zip)
    lat = "miss"
    longi = "miss"
    zip = "miss"
  lat_list.append(lat)
  long_list.append(longi) 
```

```python
attr_group =[]

for index , row in df.iterrows():
  if( 0 <=row['attrition_percentage'] < 10 ):
    attr_group.append(0)
  elif( 10 <=row['attrition_percentage'] < 20 ):
    attr_group.append(1)
  elif(20 <=row['attrition_percentage'] < 30 ):
    attr_group.append(2)
  elif( 30 <=row['attrition_percentage'] < 40 ):
    attr_group.append(3)
  elif( 40 <=row['attrition_percentage'] < 50 ):
    attr_group.append(4)
  elif( 50 <=row['attrition_percentage'] < 60 ):
    attr_group.append(5)

df['attr_group'] = attr_group

```

```python
ruca_group = []

for index , row in df.iterrows():
  if( 0 <= row['ruca'] < 4):
    ruca_group.append('urban')
  else:
    ruca_group.append('rural')
df['ruca_group'] = ruca_group
```

```python
df_zip_attr_mean=df.groupby(['zip'])['attrition_percentage'].mean().sort_values(ascending=False)
print(df_zip_attr_mean.head(100))
```

```python
df_zip_attr_mean = df_zip_attr_mean.to_frame().reset_index()
```

```python
rc = []
lat = []
longi = []
for index , row in df_zip_attr_mean.iterrows():
  zip = row['zip']
  val=  df.loc[df['zip'] == zip, 'ruca'].values[0]
  lat_n =  df.loc[df['zip'] == zip, 'latitude'].values[0]
  longi_n = df.loc[df['zip'] == zip, 'longitude'].values[0]
  #print(zip)
  #print(val)
  longi.append(longi_n)
  lat.append(lat_n)
  rc.append(val)
```

```python
df_zip_attr_mean['ruca'] = rc
df_zip_attr_mean['latitude'] = lat
df_zip_attr_mean['longitude'] = longi
```

```python

df_sub_zip_rev = df_zip_attr_mean.iloc[::-1]
df_concat = pd.concat([df_zip_attr_mean[:100], df_sub_zip_rev[:100] ], axis=0)
```

```python
m = folium.Map(location=[39.532644, -100.327413], zoom_start=4, tiles="Stamen Terrain")

for index , row in df_concat.iterrows():  

  color=""
  if(row['ruca'] <= 4.0):
    color="blue"
  else:
    color = "red"
  if(row["attrition_percentage"] > 30):
    folium.Marker(
        [row['latitude'], row['longitude']], 
        popup="<i>" +"</i>" + "<br>" + str(row['attrition_percentage']), 
        tooltip="zip:  " +row['zip'] +"     ruca:  " + str(row['ruca']),
        icon=folium.Icon(color=color, prefix='fa',icon='remove')
    ).add_to(m)
  else:
     folium.Marker(
        [row['latitude'], row['longitude']], 
        popup="<i>" +"</i>" + "<br>" + str(row['attrition_percentage']), 
        tooltip="zip:  " +row['zip'] +"     ruca:  " + str(row['ruca']),
        icon=folium.Icon(color=color, prefix='fa',icon='arrow-up')
    ).add_to(m)
m
```
